```python
# untuk mengatasi missing value
for column in df.columns:
    if df[column].dtype == 'object':
        # Jika kolom bertipe object, isi dengan mode
        df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
# mengecek kembali missing value
df.isna().sum()
```

```python
df.info()
```

```python
check_duplicate = df.duplicated().sum()

print(f"Jumlah data yang duplikat = {check_duplicate}")
```

```python
# Handling duplicate
df = df.drop_duplicates()
```

```python
# mengecek duplicate setelah di-handle
handle_duplicate = df.duplicated().sum()

print(f"Jumlah data yang duplikat = {handle_duplicate}")
```

```python
# Handling Outlier
# cek kolom numerik
df.select_dtypes(include=['float64', 'int64']).columns
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

batas_bawah = Q1 - 1.5 * IQR
batas_atas = Q3 + 1.5 * IQR

# melihat data outlier
outlier = df[(df['price'] < batas_bawah) | (df['price'] > batas_atas)]
outlier
# mengecek jumlah outlier
print("Jumlah outlier pada kolom price:", len(outlier))
```

```python
# menangani Outlier pada kolom price
df = df[(df['price'] >= batas_bawah) & (df['price'] <= batas_atas)]
```

```python
# mengecek kembali jumlah outlier pada kolom price
outlier = df[(df['price'] < batas_bawah) | (df['price'] > batas_atas)]
print("Jumlah outlier pada kolom price:", len(outlier))
```

```python
# Encoding
# mengecek kolom kategorikal
df.select_dtypes(include=['object']).columns
```

```python
df = pd.get_dummies(df, columns=['name', 'description', 'make', 'model', 'engine', 'fuel',
       'transmission', 'trim', 'body', 'exterior_color', 'interior_color',
       'drivetrain'], drop_first=True)
```
